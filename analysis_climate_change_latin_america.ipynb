Analysis of CO₂ emissions, mean temperature anomalies, and extreme climate events across the Ibero-American region.

````python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns
import numpy as np

plt.rcParams['figure.dpi'] = 120
sns.set_theme(style='whitegrid')
print('Libraries loaded ✓')
````

## 1. CO₂ Emissions per Capita (tonnes, 2000–2024)

````python
data_co2 = {
    'Country': ['Spain','Mexico','Brazil','Argentina','Colombia','Chile','Peru'],
    '2000': [8.8, 3.7, 2.3, 3.9, 1.6, 4.3, 1.4],
    '2005': [9.1, 4.2, 2.5, 4.5, 1.9, 4.7, 1.7],
    '2010': [7.8, 4.0, 2.7, 4.8, 2.0, 4.6, 1.9],
    '2015': [6.4, 3.8, 2.6, 4.5, 2.1, 4.8, 2.1],
    '2020': [5.1, 3.2, 2.2, 4.0, 1.9, 4.5, 2.0],
    '2024': [4.8, 3.5, 2.4, 4.3, 2.2, 4.9, 2.3]
}
np.random.seed(42)
df_co2 = pd.DataFrame(data_co2).set_index('Country')
print(df_co2)
````

````python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

ax1 = axes[0]
years = ['2000','2005','2010','2015','2020','2024']
for country in df_co2.index:
    ax1.plot(years, df_co2.loc[country, years], marker='o', linewidth=2.5, label=country)
ax1.set_title('CO₂ Emissions per Capita\n(tonnes, 2000–2024)', fontsize=13, fontweight='bold')
ax1.set_xlabel('Year')
ax1.set_ylabel('Tonnes CO₂ per capita')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Renewable energy share
ax2 = axes[1]
renewables = {
    'Country': ['Spain','Mexico','Brazil','Argentina','Colombia','Chile','Peru'],
    '2010': [30.2, 14.4, 44.8, 9.8, 67.2, 31.2, 61.4],
    '2024': [52.1, 28.6, 83.4, 18.2, 72.1, 54.8, 72.3]
}
df_ren = pd.DataFrame(renewables).set_index('Country')
ax2.barh(df_ren.index, df_ren['2024'], color=sns.color_palette('Greens_r', n_colors=7), alpha=0.85)
for i, (idx, row) in enumerate(df_ren.iterrows()):
    ax2.text(row['2024'] + 0.5, i, f'{row["2024"]:.1f}%', va='center', fontsize=10)
ax2.set_title('Renewable Energy Share (2024)', fontsize=13, fontweight='bold')
ax2.set_xlabel('% of total electricity generation')
ax2.grid(True, alpha=0.3, axis='x')

plt.tight_layout()
plt.savefig('climate_change_iberoamerica.png', dpi=150, bbox_inches='tight')
plt.show()
````
